# Using [Annoy](https://github.com/spotify/annoy) for Approximate Nearest Neigbors

Annoy is a convenient tool built by Spotify to build solutions for Approximate Nearest Neighbors.  ANN helps keep inference times quick when building out movie nearest neighbors

```python
from annoy import AnnoyIndex
import pandas as pd
```

```python
ratings = pd.read_parquet('training_data/movielens_25m_5_5.pq')[['USER_ID', 'MOVIE_ID', 'RATING']]
```

###  Nearest neighbors based on historical user interactions

```python
util_df=pd.pivot_table(data=ratings,values='RATING',index='USER_ID',columns='MOVIE_ID').fillna(0)
util_df
```

### Train the model

```python
from tqdm import tqdm
t = AnnoyIndex(util_df.shape[0], 'angular')
for i in tqdm(util_df):
    v = util_df[i]
    t.add_item(int(i), v)
```

```python
t.build(10) # 10 trees
```

### Save the model for inference later

```python
t.save('movies.ann')
```
